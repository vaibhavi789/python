## Inheritance 

When a child class (***Derived class) inherits the properties of a parent class (***Base class)

class Parent:
    def function1(self):
        print("This is a parent function")

class Child(Parent):
    def function2(self):
        print("This is a child function")

```python
c = Child()
```

```python
c.function1()
```

```python
c.function2()
```

```python
class Animals:   ###Parent class
    def eat(self):
        print('I can eat')
        
class Cat(Animals):  ### Child class
    def grumpy(self):
        print('I am grumpy')
class Dog(Animals):
    def bark(self):
        print('I bark')
```

```python
dog1=Dog()
```

```python
dog1.eat()
```

```python
dog1.bark()
```

```python
cat1=Cat()
```

```python
cat1.eat()
```

__init__ function: (Initialization constructor)

Init function is called automatically everytime the class is used to create a object

```python
class Polygon:
    def __init__(self, sides):
        self.sides = sides    #### public variables
    def display_info(self):
        print("A polygon is a 2-D with straight lines")
    def get_perimeter(self):
        perimeter = sum(self.sides)
        return perimeter
    

class triangle(Polygon):
    def display_info(self):
        print("A triangle has 3-sides")

class quadrilateral(Polygon):
    def display_info(self):
        print("A quadrilateral has 4-sides")
```

```python
t=triangle([5,6,8])
```

```python
perimeter=t.get_perimeter()
print("The perimeter of the triangle is", perimeter)
```

 #### method overriding
 
 It can be achieved to change the functionality of parent class function

```python
t.display_info()  ## method overriding
```

Types of Inheritance:


1. Single - Involves 1 child class and 1 parent class

2. Multiple - Involves more than 1 parent class

3. Multi-level -  child class acts as a parent class for another child class

4. Hierarchical- More than 2 derived classes from a single base class 

5. Hybrid - More than 1 type of inheritance

```python
class Parent:
    def fn1(self):
        print('This is function1')

class Parent2(Parent):
    def fn3(self):
        print('This is function3')

class Parent3(Parent):
    def fn4(self):
        print('This is function4')
        
class Child(Parent2):
    def fn2(self):
        print('This is function2')
```

```python
ob=Child()
ob.fn3()
```

```python
class Car:
    def __init__(self, windows, doors, engine):
        self.windows = windows  ### self._windows- protected variable,
        self.doors = doors      ### self.__windows- private variable,
        self.engine = engine
    def drive(self):
        print('I love driving')
        
class audi(Car):
    def __init__(self, windows, doors, engine, voice):
        self.voice = voice
    def selfdriving(self):
        print('Audi supports self-driving')
    
```

```python
a7=audi(4,5,'diesel',True)
a7.drive()
```

```python
car1=Car(2, 3, 'petrol')
car1.engine
```
